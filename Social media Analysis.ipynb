```python
x = ["Mobile", "Laptop", "Nothing"]
y=[59, 3, 1]

plt.bar(x, y, color=['r', 'g','b','brown','hotpink'], edgecolor="black")
plt.title('most used devices for social media', fontsize=16)
plt.ylabel("count", fontsize=13)
plt.xlabel("device name", fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
style.use('fivethirtyeight')
plt.figure(figsize=(10,7))
plt.show()             
```

```python
plt.style.use("fivethirtyeight")
slices = [59, 3, 1]
labels = ["Mobile", "Laptop", "Nothing"]
explode=(0.1, 0,0.2)
plt.pie(slices, explode=explode,labels=labels, autopct='%.2f%%', shadow=True, startangle=90)
plt.title('Most device used')
plt.axis('equal')
plt.legend(loc='upper right')
plt.style.use('ggplot')
plt.show()
```
